```python
import pickle

with open("outputs_100.pkl", "rb") as f:
    outputs = pickle.load(f)
```

```python
import json
input_path = "/local/huzaifa/workspace/vLLM/vllm-serverless-optimization/new_idea/traces/modified_conversation_trace_100.jsonl"
output_path = "/local/huzaifa/workspace/vLLM/vllm-serverless-optimization/new_idea/traces/modified_conversation_trace_100_with_proc_time.jsonl"
with open(input_path, "r") as f:
    lines = []
    for i, line in enumerate(f):
        line = line.strip()
        if not line:
            continue
        obj = json.loads(line)
        obj["generation_time"] = outputs[i].latency
        lines.append(obj)

with open(output_path, "w") as fout:
    for obj in lines:
        fout.write(json.dumps(obj) + "\n")
```
